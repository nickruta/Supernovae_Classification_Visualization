# Supernova Photometric Classification

[The original paper.](https://arxiv.org/abs/1001.5210)

A supernovae (SN) is the explosion of a massive star at the end of its evolutionary life. This explotion produces a huge amount of light making even far away stars visible by us through a relatively short period of time. Astronomers have classified SNe in terms of their light curves and spectral absortion lines. A rough classification scheme can be described as follows:
* **Type I**: Show no Hydrogen lines
 * **Type Ia**: Show a strong ionised silicon absorption line at 615.0 nm.
 * **Type Ib**: Shows a non-ionized helium (He I) line at 587.6 nm
 * **Type Ic**: Weak or no helium.
* **Type II**: Show Hydrogen lines

Spectra are expensive, so light curves of SNe are used to classify them.

<img src="https://upload.wikimedia.org/wikipedia/commons/e/e0/Comparative_supernova_type_light_curves.png" width="50%">

The **Supernova Photometric Classification Challenge** have publicly released a mix of simulated SNe, with types (Ia, Ib, Ic, II) selected in proportion to their expected rate.

## The Data

Each supernovae is described by:
* An _ID_
* Its coordinates _RA_ and _DEC_
* Its dust extinction _MWEBV_
* Its _redshift_
* Its _type_
 * Ia: 1
 * II: 2
 * Ibc: 3
 * IIn: 21
 * IIP: 22
 * IIL: 23
 * Ib: 32
 * Ic: 33

Each SN has observations in different filters: g, r, i, and z

 <img src="http://des-docdb.fnal.gov/cgi-bin/RetrieveFile?docid=4295&filename=DES_100mmx100mm_grizy_trans.jpg&version=2" width="50%">
 
These are represented in their respective light curves by:
* Modified Julian Date _MJD_
* Magnitude
* Magnitude error

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv ('data/SNe.csv')
df.head(10)
```

```python
print (df['type'].unique())
```

```python
df_TypeI = df[df['type'] == 1]
df_TypeI.head()
```

```python
t = df_TypeI[df_TypeI['ID'] == 642]['MJD']
g = df_TypeI[df_TypeI['ID'] == 642]['g']
g_err = df_TypeI[df_TypeI['ID'] == 642]['g_err']
```

```python
plt.clf()
plt.errorbar (t, g, yerr = g_err)
plt.show()
```

```python
id_max = df.groupby(['ID'])['g'].transform(max) == df['g']
df_max = df[id_max]
df_max[df_max['ID'] == 642].head()
```

```python
# Let's check that out
df_TypeI[df_TypeI['ID'] == 642]['g'].max()
```

```python
# Calculate SNR at peak

snr = df_max['g']/df_max['g_err']
df_max['SNR'] = snr
df_max.head(10)
```

```python
sorted_df = df_max.sort_values(by=['SNR'], ascending = False)
sorted_df.head(10)


```

```python
print (sorted_df['ID'].iloc[0])
t = df[df['ID'] == sorted_df['ID'].iloc[1]]['MJD']
g = df[df['ID'] == sorted_df['ID'].iloc[1]]['g']
g_err = df[df['ID'] == sorted_df['ID'].iloc[1]]['g_err']

plt.clf()
plt.errorbar (t, g, yerr = g_err)
plt.show()
```

```python
# print 10 brightest SNe
N = 10

plt.clf()
for i in range(N):
    t = df[df['ID'] == sorted_df['ID'].iloc[i]]['MJD']
    g = df[df['ID'] == sorted_df['ID'].iloc[i]]['g']
    g_err = df[df['ID'] == sorted_df['ID'].iloc[i]]['g_err']

    plt.errorbar (t, g, yerr = g_err, label = "Type " + str(sorted_df['type'].iloc[i]))
plt.legend(loc = "upper right")
plt.show()
```
